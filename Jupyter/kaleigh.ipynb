## Animating the yield surface to spot errors

```python
from celluloid import Camera ## Module required for animating figures
def YieldSurface(km, p_i, M_i, M_tc, Chi_tc, psi_i, lambda_c, ps, N):
    CHIi=Chi_tc/(1-Chi_tc*lambda_c/M_tc)
    M_itc=M_tc*(1-CHIi*N*abs(psi_i)/M_tc)
    p_max=p_i/np.exp(-CHIi*psi_i/M_itc)
    pr=p_max*(1+km)
    pl=p_max*(1-km)
    n_L=M_i*(1-CHIi*psi_i/M_itc)
    qs=[]
    for p in ps:        
        if (p<=pr):
            q=-p*M_i*(1+np.log(p_i/p))
        elif (p<=pl):
            sigma=1+km
            iota=1-km
            C_1=(np.log(sigma)-km)/(4*km**3)
            C_2=((np.log(sigma)+1)/(2*(sigma-iota)))+3*C_1
            C_3=-3*(sigma**2)*C_1+2*sigma*C_2-np.log(sigma)-1
            C_4=-(sigma**3)*C_1+(sigma**2)*C_2-sigma*(C_3+np.log(sigma))
            q=-(M_i*C_1*p**3/(p_max**2))+(M_i*C_2*p**2/p_max)-((n_L+M_i*C_3)*p)-(p_max*M_i*C_4) 	
        else:
            q=-n_L*p
        qs.append(q)
        
    return qs
                
    
```

```python
from celluloid import Camera
import numpy as np
from IPython.display import HTML
fig = plt.figure()
camera = Camera(fig)
for i in range(len(NorSand_VT_data[0])):
    ps=NorSand_VT_data[0]['p'][:i]
    qs=NorSand_VT_data[0]['q'][:i]
    p_i=NorSand_VT_data[0]['statev(6)'][i] #Image pressure from benchmark
    M_i=NorSand_VT_data[0]['statev(7)'][i] #Image stress ratio from benchmark
    Chi_tc=NorSand_VT_data[0]['statev(5)'][i] #Dilatancy state parameter
    psi_i=NorSand_VT_data[0]['statev(4)'][i] # state parameter
    # Plot first yield surface
    km=0.1
    p_intersec=np.exp(1)*p_i
    ps_s=np.linspace(0, p_intersec, 100)
    qs_s=YieldSurface(km, p_i, M_i, 1.27, Chi_tc, psi_i, 0.021, ps_s, 0.4)
    plt.plot(ps_s, qs_s, '-', color='b')
    plt.plot(ps,qs,'-',color='r')
    
    camera.snap()
animation = camera.animate()
HTML(animation.to_html5_video())
```
